### Logo, Title

```python
#function for displaying dataframes

def setup_ui(df):

    out = widgets.Output()
    with out:
        display(df)
    return out
```

```python
#nba logo and page title

file = open("nba.png", "rb")
image = file.read()

image_headline = widgets.Image(
                    value=image,
                    format='jpg',
                    width='300',
                    layout=widgets.Layout(margin='auto')
                )

# label_headline = widgets.Label(
#                     value='NBA Player Stats 2019-2020',
#                     style={'description_width': 'initial'},
# #                     layout= widgets.Layout(width='50%', height='80px')
#                 )

title = widgets.HTML(value = "<h1>NBA Player Stats 2019-2020</h1>", layout=widgets.Layout(margin='auto'))

vbox_headline = widgets.VBox([image_headline, title])
```

### Dislay Stats by Searching Player Name

```python
# player query text box

player_title = widgets.HTML(value = "<h3>Search for Player:</h3>")

player = widgets.Text(placeholder='Enter Player Name')
```

```python
# sort results widget

sortby1 = widgets.Dropdown(
    options=col_names,
    value='Rk',
    description='Sort by:',
    disabled=False,
)
```

```python
player_sort = widgets.HBox([player, sortby1])
```

```python
#button to display results by querying by player name

button_search = widgets.Button(
                description='Search Player Stats',
                style={'description_width': 'initial'}
            )

output = widgets.Output()

def on_button_clicked(event):
    with output:
        clear_output()
        print(f"Player: {player.value}")
        player_input = 'SELECT * FROM PlayerStats WHERE Player LIKE "%' + player.value + '%" ORDER BY ' + sortby1.value + ' ASC;'
        pd.set_option("display.max_rows", None, "display.max_columns", None)
        playerstats_df = pd.read_sql_query(player_input, mydb)
        print("Number of Results:", len(playerstats_df))
        player_box = setup_ui(playerstats_df)
        player_box_df = widgets.VBox([player_box])
        display(player_box_df)
        
        
button_search.on_click(on_button_clicked)

vbox_result = widgets.VBox([button_search, output])
```

### Display Stats by Team

```python
#select by category (team and count)

team_title = widgets.HTML(value = "<h3>View Stats by Team:</h3>")

select_team = widgets.Select(
    options=teamcnt_list,
    value='All',
    style={'description_width': 'initial'},
#     description='OS:',
    disabled=False
)
```

```python
# sort results widget

sortby2 = widgets.Dropdown(
    options=col_names,
    value='Tm',
    description='Sort by:',
    disabled=False,
)
```

```python
team_sort = widgets.HBox([select_team, sortby2])
```

```python
#display results by category (team)

button_team = widgets.Button(
                description='Display Team Stats',
                style={'description_width': 'initial'}
            )

output_team = widgets.Output()

def on_team_button_clicked(event):
    with output_team:
        clear_output()
        print(f"Team: {select_team.value[0:3]}")
        if (select_team.value[0:3] == "All"):
            team_input = 'SELECT * FROM PlayerStats ORDER BY ' + sortby2.value + ' ASC;' 
            pd.set_option("display.max_rows", None, "display.max_columns", None)
            teamstats_df = pd.read_sql_query(team_input, mydb)
            print("Number of Results:", len(teamstats_df))
            team_box = setup_ui(teamstats_df)
            team_box_df = widgets.VBox([team_box])
            display(team_box_df)
        else:
            team_input = 'SELECT * FROM PlayerStats WHERE Tm = "' + select_team.value[0:3] + '" ORDER BY ' + sortby2.value + ' ASC;'
            pd.set_option("display.max_rows", None, "display.max_columns", None)
            teamstats_df = pd.read_sql_query(team_input, mydb)
            print("Number of Results:", len(teamstats_df))
            team_box = setup_ui(teamstats_df)
            team_box_df = widgets.VBox([team_box])
            display(team_box_df)
        
        
button_team.on_click(on_team_button_clicked)

vbox_team_result = widgets.VBox([button_team, output_team])
```

### Display Stats by Player Position

```python
#select by category (position and count)

pos_title = widgets.HTML(value = "<h3>View Stats by Player Position:</h3>")

select_pos = widgets.Select(
    options=poscnt_list,
    value='All',
    style={'description_width': 'initial'},
    disabled=False
)
```

```python
# sort results widget

sortby3 = widgets.Dropdown(
    options=col_names,
    value='Pos',
    description='Sort by:',
    disabled=False,
)
```

```python
pos_sort = widgets.HBox([select_pos, sortby3])
```

```python
#display results by category (position)

button_pos = widgets.Button(
                description='Display Position Stats',
                style={'description_width': 'initial'}
            )

output_pos = widgets.Output()

def on_pos_button_clicked(event):
    with output_pos:
        clear_output()
        
        if (select_pos.value[0:3] == "All"):
            print("Position: All")
            pos_input = 'SELECT * FROM PlayerStats ORDER BY ' + sortby3.value + ' ASC;' 
            pd.set_option("display.max_rows", None, "display.max_columns", None)
            posstats_df = pd.read_sql_query(pos_input, mydb)
            print("Number of Results:", len(posstats_df))
            pos_box = setup_ui(posstats_df)
            pos_box_df = widgets.VBox([pos_box])
            display(pos_box_df)
        else:
            print(f"Position: {select_pos.value.partition('(')[0].strip()}")
            pos_input = 'SELECT * FROM PlayerStats WHERE Pos = "' + select_pos.value.partition('(')[0].strip() + '" ORDER BY ' + sortby3.value + ' ASC;'
            pd.set_option("display.max_rows", None, "display.max_columns", None)
            posstats_df = pd.read_sql_query(pos_input, mydb)
            print("Number of Results:", len(posstats_df))
            pos_box = setup_ui(posstats_df)
            pos_box_df = widgets.VBox([pos_box])
            display(pos_box_df)
        
        
button_pos.on_click(on_pos_button_clicked)

vbox_pos_result = widgets.VBox([button_pos, output_pos])
```
